```python
search_texts = []
metadata = []

for table in input_schema:
    for column in table["columns"]:
        search_texts.append(f"Table: {table['table']}, Column: {column['name']}, Description: {column['description']}")
        metadata.append({"table": table["table"], "column": column["name"]})
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Load pre-trained embedding model
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")

# Function to generate embeddings
def generate_embeddings(texts):
    inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
        embeddings = outputs.last_hidden_state.mean(dim=1)  # Mean pooling
    return embeddings

embeddings = generate_embeddings(search_texts)
```

```python
import faiss

# Initialize FAISS index
d = embeddings.shape[1]  # Dimension of embeddings
index = faiss.IndexFlatL2(d)
index.add(embeddings.numpy())  # Add embeddings to the index
```

```python
def search_faiss(query, k=10):
    query_embedding = generate_embeddings([query]).numpy()
    distances, indices = index.search(query_embedding, k)  # Perform search
    for i, idx in enumerate(indices[0]):
      print(f"Rank {i + 1}:")
      print(f"Description: {metadata[idx]}")
      #print(f"Type: {metadata[idx]}, Name: {metadata[idx]}")
      print(f"Distance: {distances[0][i]}\n")

    results = [metadata[i] for i in indices[0]]  # Retrieve corresponding metadata
    # for i in indices:
    #     results[i]['distance'] = distances[0][i]  # Add distance to metadata
    return results

# Example search
query = "Find department name from detartments"
results = search_faiss(query)
print(results)
```

# Storing table and column in hierarchy

```python
table_texts = [f"Table: {table['table']}, Description: {table['description']}" for table in input_schema]
column_texts = [
    {
        "table": table["table"],
        "column_texts": [
            f"Table: {table['table']}, Column: {col['name']}, Description: {col['description']}"
            for col in table["columns"]
        ]
    }
    for table in input_schema
]
```

```python
from transformers import AutoTokenizer, AutoModel
import faiss
import torch

# Load embedding model
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")

def generate_embeddings(texts):
    inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
        embeddings = outputs.last_hidden_state.mean(dim=1)  # Mean pooling
    return embeddings.numpy()

# Generate table embeddings
table_embeddings = generate_embeddings(table_texts)

# Create FAISS index for tables
table_index = faiss.IndexFlatL2(table_embeddings.shape[1])
table_index.add(table_embeddings)

# Generate column embeddings and create a separate index for each table
column_indexes = {}
for table_data in column_texts:
    column_embeddings = generate_embeddings(table_data["column_texts"])
    column_index = faiss.IndexFlatL2(column_embeddings.shape[1])
    column_index.add(column_embeddings)
    column_indexes[table_data["table"]] = {"index": column_index, "metadata": table_data["column_texts"]}
```

```python
def hierarchical_search(query, tables, column_indexes, top_k_tables=10, top_k_columns=3):
    # Step 1: Search the table index
    query_embedding = generate_embeddings([query])
    _, table_indices = table_index.search(query_embedding, top_k_tables)

    # Retrieve top tables
    selected_tables = [tables[i] for i in table_indices[0]]
    print("Top Tables:", [t["table"] for t in selected_tables])

    # Step 2: Search within columns of the selected tables
    column_results = []
    for table in selected_tables:
        table_name = table["table"]
        column_index = column_indexes[table_name]["index"]
        column_metadata = column_indexes[table_name]["metadata"]

        _, column_indices = column_index.search(query_embedding, top_k_columns)
        column_results.extend([column_metadata[i] for i in column_indices[0]])

    return column_results
```

```python
# Example query
query = "Find department and employees working in the departments"
results = hierarchical_search(query, input_schema, column_indexes)
print("Top Columns:", results)
```
